```python
import networkx as nx
import matplotlib.pyplot as plt

# Buat graph kosong
G = nx.Graph()

# Tambahkan 9 node
G.add_nodes_from(range(1, 10))

# Tambahkan edges sesuai dengan pola pada gambar
edges = [
    (1, 4), (1, 3), (1, 2),
    (2, 1), (2, 3),
    (3, 4), (3, 1), (3, 2),
    (4, 5), (4, 6),
    (5, 6), (5, 7), (5, 8),
    (6, 7), (6, 8),
    (7, 9), (7, 8)
]
G.add_edges_from(edges)

# Gambar graph
plt.figure(figsize=(8, 8))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=700, font_size=12, font_color='black', font_weight='bold')
plt.title("Graph with 9 Nodes")
plt.show()
```

```python
# Hitung degree centrality
degree_centrality = nx.degree_centrality(G)

# Tampilkan hasil degree centrality masing-masing node
for node, centrality in degree_centrality.items():
    print(f"Node {node}: Degree Centrality = {centrality}")
```

```python
# Hitung closeness centrality
closeness_centrality = nx.closeness_centrality(G)

# Tampilkan hasil closeness centrality masing-masing node
print("\nCloseness Centrality:")
for node, centrality in closeness_centrality.items():
    print(f"Node {node}: Closeness Centrality = {centrality:.4f}")
```

```python
# Hitung betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Tampilkan hasil betweenness centrality masing-masing node
print("\nBetweenness Centrality:")
for node, centrality in betweenness_centrality.items():
    print(f"Node {node}: Betweenness Centrality = {centrality:.4f}")
```
